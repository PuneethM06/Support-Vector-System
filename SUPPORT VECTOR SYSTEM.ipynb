# SUPPORT VECTOR MACHINE

IN THIS TUTORIAL BASED ON PETAL LENGTH WE DECIDE WHETER THE FLOWER IS SETOSA OR VERSICOLOR

SUPPORT VECTOR MACHINE TRIES TO DRAW THE LINE WITH HIGHER MARGIN TO CLASSIFY TWO GROUPS BASED ON DATA POINTS FOR BETTER CLASSIFICATION

```python
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

CREATING A DF


```python
df = pd.DataFrame(iris.data,columns = iris.feature_names)
df
```

```python
df['target'] = iris.target
df
```

```python
iris.target_names
```

TARGET NAMES; 0 = SETOSA, 1= VERSICOLOR , 2 = VIRGINICA

```python
df[df.target==1].head()
```

```python
df[df.target==0].head()
```

FROM 0-50 IT IS SETOSA , 50-100 IT IS VERSICOLOR AND THE REMAINING 50 IS THE OTHER

CREATING ONE COLUMN FROM ANOTHRE COLUMN OR ALREADY EXISTING COLUMN

```python
df['flower_name'] = df.target.apply(lambda x :iris.target_names[x] )
df
```

DATA VISUALISATION

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

SEPERATING 3 DIFFERENT DFFOR 3 DIFFERENTS DATA

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
df1.head()
```

```python
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color = 'blue',marker = '*')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color = 'green',marker = '*')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
```

```python
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color = 'blue',marker = '*')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color = 'green',marker = '*')
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
```

TRAINING THE MODEL

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target','flower_name'],axis = 1)
X.head()
```

```python
Y = df.target
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2)
```

IMPORTING SUPPORT VECTOR SYSTEM

```python
from sklearn.svm import SVC 
model = SVC(C = 10)
```

C IS AN OPTIONAL PARAMETER N IT REPRESENTS THE REGULARIZATION

```python
model.fit(X_train,Y_train)
```

THE MODEL ACCURACY IS 93 %

```python
model.score(X_test,Y_test)
```

```python

import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
digits.target
```

```python
dir(digits)
```

```python
digits.target_names
```

```python
df = pd.DataFrame(digits.data,digits.target)
df.head()
```

```python
df['target'] = digits.target
df.head(20)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('target',axis='columns'), df.target, test_size=0.3)
```

```python

from sklearn.svm import SVC
rbf_model = SVC(kernel='rbf')
```

```python

rbf_model.fit(X_train, y_train)
```

```python

rbf_model.score(X_test,y_test)
```

# USING LINEAR MODEL

```python

linear_model = SVC(kernel='linear')
linear_model.fit(X_train,y_train)
```

```python

linear_model.score(X_test,y_test)
```
